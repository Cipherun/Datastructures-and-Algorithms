# Removing given character from string using recursion

```python
def remove_character(s:str,ch:str)->str:
    if (len(s)==0 or s==""):
        return s
    ans=remove_character(s[1:],ch)
    if s[0]==ch:
        return ans
    else:
        return s[0]+ans
string="hello goodzzz morning"
print(remove_character(string,'z'))
```

```python
# reversing an array
def reverse(arr:list[int],left:int,right:int):
    if len(arr)==0 or len(arr)==1:
        return arr
    if left<=right:
        
        arr[left],arr[right]=arr[right],arr[left]
        
        return reverse(arr,left+1,right-1)
        
    else:
        return arr
print(reverse([1,2,3],0,len([1,2,3])-1))
    
```

```python
# palindrome checking
def is_palindrome(s:str,left:int,right:int)->bool:
    
    if s=='':
        return True
    if len(s)==1:
        return True
    if left<right and s[left]==s[right]:
        return is_palindrome(s,left+1,right-1)
    elif left==right:
        return True
    else:
        return False
print(is_palindrome('naman',0,len('naman')-1))
    
    
```

```python
# subsequences printing on array
def print_sub(ind:int,arr,n,lst,k):
    if ind==n:
        if sum(lst)==k:

            print(lst)
        # if len(lst)==0:
        #     print('[]')
        return
    lst.append(arr[ind])
    print_sub(ind+1,arr,n,lst,k)
    lst.pop()

    print_sub(ind+1,arr,n,lst,k)
    
if __name__=='__main__':
    print_sub(0,[1,2,4,5,7],len([1,2,4,5,7]),[],7)
    
```
